```python
import random

class RailwayTicketBooking:
    datelist=[]
    bookingDetails={}

    def train_selection(self):
        print("\nWhich train would you like to book?\n1. Rajdhani Express\n2. Shatabdi Express\n3. Duronto Express\n4. Exit")
        return int(input())

    def station_selection(self, train_choice):
        print("\nSelect a station:\n1. New Delhi\n2. Mumbai Central\n3. Kolkata Howrah\n4. Chennai Central\n5. Exit")
        return int(input())

    def coach_selection(self, train_choice):
        print("\nSelect a coach:\n1. AC First Class\n2. AC 2 Tier\n3. AC 3 Tier\n4. Sleeper Class\n5. Exit")
        return int(input())

    def seat_selection(self):
        print("\nSelect number of tickets:")
        return int(input())

    def generate_booking_code(self):
        return random.randrange(1000, 9999)

    def book_ticket(self, train_choice, station_choice, coach_choice, seat_count):
        date = input("\nEnter date (MM/DD/YYYY): ")
        name = input("Enter passenger's name: ")
        phone = input("Enter passenger's phone number: ")
        booking_code = self.generate_booking_code()
        details = [date, name, phone, train_choice, station_choice, coach_choice, seat_count]
        self.bookingDetails[booking_code] = details
        print("\nYour booking details:")
        print("Booking Code:", booking_code)
        print("Train:", train_choice)
        print("Station:", station_choice)
        print("Coach:", coach_choice)
        print("Date:", date)
        print("Passenger's Name:", name)
        print("Passenger's Phone Number:", phone)
        print("Number of Tickets:", seat_count)

    def cancel_booking(self):
        booking_code = int(input("Enter booking code to cancel: "))
        if booking_code in self.bookingDetails:
            del self.bookingDetails[booking_code]
            print("Booking canceled successfully.")
        else:
            print("Booking not found.")

    def display_bookings(self):
        print("\nBooking Details:")
        for code, details in self.bookingDetails.items():
            print("Booking Code:", code)
            print("Train:", details[3])
            print("Station:", details[4])
            print("Coach:", details[5])
            print("Date:", details[0])
            print("Passenger's Name:", details[1])
            print("Passenger's Phone Number:", details[2])
            print("Number of Tickets:", details[6])
            print()

railway_booking = RailwayTicketBooking()

while True:
    print("\nWelcome to Railway Ticket Booking System\nPlease select an option:")
    print("1. Book Ticket\n2. Cancel Booking\n3. View Bookings\n4. Exit")
    choice = int(input())

    if choice == 1:
        train_choice = railway_booking.train_selection()
        if train_choice == 4:
            continue
        station_choice = railway_booking.station_selection(train_choice)
        if station_choice == 5:
            continue
        coach_choice = railway_booking.coach_selection(train_choice)
        if coach_choice == 5:
            continue
        seat_count = railway_booking.seat_selection()
        railway_booking.book_ticket(train_choice, station_choice, coach_choice, seat_count)

    elif choice == 2:
        railway_booking.cancel_booking()

    elif choice == 3:
        railway_booking.display_bookings()

    elif choice == 4:
        print("Thank you for using Railway Ticket Booking System. Have a great journey!")
        break

    else:
        print("Invalid choice. Please try again.")
```
